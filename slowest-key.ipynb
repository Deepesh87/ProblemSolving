```python
def slowestKey(releaseTimes,keysPressed):
    hash={}
    ans=''
    max_=releaseTimes[0]

    for i in range(len(keysPressed)):
        if i==0:
            hash[keysPressed[i]]=releaseTimes[i]
        else:
            if keysPressed[i] not in hash:
                hash[keysPressed[i]]=releaseTimes[i]-releaseTimes[i-1]
                if hash[keysPressed[i]]>max_:
                    max_=hash[keysPressed[i]]
            else:
                if releaseTimes[i]-releaseTimes[i-1]>hash[keysPressed[i]]:
                    hash[keysPressed[i]]=releaseTimes[i]-releaseTimes[i-1]
                    if hash[keysPressed[i]]>max_:
                        max_=hash[keysPressed[i]]


    for i in hash:
        if hash[i]==max_:
            ans+=i
    return sorted(ans)[-1]

#O(N) time and O(1)space, the hash table will have 26 characters max. But i am sorting some characters so not 
#exactly O(N)time plus storing a few characters in ans
```
